```python
import numpy as np
from astropy.io import fits
import os

# Directory containing the .npy files
kappa_dir = '/share/gpu0/jjwhit/samples/'

for i in range(1, 21):
    kappa_npy = os.path.join(kappa_dir, f'np_gt_{i}.npy')
    kappa_fits = os.path.join(kappa_dir, f'fits_gt_{i}.fits')

    # Load the .npy file
    npy_data = np.load(kappa_npy)
    real_part = npy_data.real
    imag_part = npy_data.imag
    
    # Convert to FITS and save
    hdu_real = fits.ImageHDU(real_part, name='Real')
    hdu_imag = fits.ImageHDU(imag_part, name='Imag')

    hdu_primary = fits.PrimaryHDU()
    hdul = fits.HDUList([hdu_primary, hdu_real, hdu_imag])
    hdul.writeto(kappa_fits, overwrite=True)
```

```python
# Make shear maps
def realistic_noise_maker(kappa: np.ndarray) -> np.ndarray:
    std1 = np.load('/home/jjwhit/rcGAN/mass_map_utils/cosmos/cosmos_std1.npy', allow_pickle=True)
    std2 = np.load('/home/jjwhit/rcGAN/mass_map_utils/cosmos/cosmos_std2.npy', allow_pickle=True)
    gamma = MMDataTransform.forward_model(kappa, kernel) + (
        std1 * np.random.randn(300, 300) #300 = image size
        + 1.j * std2 * np.random.randn(300, 300))
    return gamma
```
